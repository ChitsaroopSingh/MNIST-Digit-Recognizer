Importing Libraries

```python
import pickle
import joblib
import time
from sklearn.metrics import accuracy_score
```

Loading QMNIST Dataset

```python
with open("dataset/qmnist.pkl", "rb") as f:
    qmnist = pickle.load(f)

X = qmnist["data"]
y = qmnist["labels"].ravel()

# Flatten 28x28 images into 784 features
X = X.reshape(-1, 28 * 28)

print(f"Dataset shape: {X.shape}")
print(f"Labels shape: {y.shape}")
```

SGD Classifier Accuracy

```python
sgd = joblib.load("models/sgd_model.joblib")

start_time = time.time()
y_pred_sgd = sgd.predict(X).astype(int)
end_time = time.time()

accuracy_sgd = accuracy_score(y, y_pred_sgd)
time_sgd = end_time - start_time

print(f"SGD Accuracy: {accuracy_sgd * 100:.2f}%")
print(f"SGD Prediction Time: {time_sgd:.2f} seconds")
```

Decision Tree Accuracy

```python
dt = joblib.load("models/dt_model.joblib")

start_time = time.time()
y_pred_dt = dt.predict(X).astype(int)
end_time = time.time()

accuracy_dt = accuracy_score(y, y_pred_dt)
time_dt = end_time - start_time

print(f"Decision Tree Accuracy: {accuracy_dt * 100:.2f}%")
print(f"Decision Tree Prediction Time: {time_dt:.2f} seconds")
```

KNN Accuracy

```python
knn = joblib.load("models/knn_model.joblib")

start_time = time.time()
y_pred_knn = knn.predict(X).astype(int)
end_time = time.time()

accuracy_knn = accuracy_score(y, y_pred_knn)
time_knn = end_time - start_time

print(f"KNN Accuracy: {accuracy_knn * 100:.2f}%")
print(f"KNN Prediction Time: {time_knn:.2f} seconds")
```
