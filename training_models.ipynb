Importing common libraries

```python
import os
import joblib
```

Fetching the MNIST Dataset

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml("mnist_784", as_frame=False)
```

Setting up the labels

```python
X = mnist.data
y = mnist.target
print(X.shape) #there are total of 70,000 and each image has 28x28 = 784 pixels
print(y.shape) #total of 70,000 images so a unique number
```

Training the SGD Classifier and checkpointing it with joblib

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X,y)

os.makedirs("models", exist_ok=True)
joblib.dump(sgd_clf, "models/sgd_model.joblib") #trained model is stored in .joblib file in a separate folder
```

Training Decision Tree Classifier Model

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(X,y)

joblib.dump(dt_clf, "models/dt_model.joblib")
```

Training SVM Classifier

```python
from sklearn.svm import LinearSVC

sv_clf = LinearSVC(random_state=42)
sv_clf.fit(X,y)

joblib.dump(sv_clf, "models/svc_model.joblib")
```

Training KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

knn_clf = KNeighborsClassifier()
knn_clf.fit(X,y)

joblib.dump(knn_clf, "models/knn_model.joblib")

```
